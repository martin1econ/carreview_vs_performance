```python
# We are going to upload three different datasets, stocks and cust_reviews(customers)/expert_reviews

## We want to find out if customer and expert reviews are significantly different
## and combine them if possible 

## We want to see how reviews and stocks move throughout the years for each Company(Make)
```

```python
#libraries needed

import pandas as pd
import numpy as np
import matplotlib as mpl
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
## From the previous step I saved with first column as index
### so add index_col=[0]

## Stocks Dataframe

stocks = pd.read_csv('cleaned_data/date_symbol_open1.csv',index_col=[0])
stocks[stocks['Symbol']=='F'].head()
```

```python
## Customer dataframe

cust_reviews = pd.read_csv('cleaned_data/review_x_symbol1.csv',index_col=[0])
cust_reviews['Reviewed_by'] = 'customer'
cust_reviews[cust_reviews['Company']=='toyota'].head()
```

```python
## Expert reviews dataframe

expert_reviews = pd.read_csv('cleaned_data/symbol_year_expert1.csv',index_col=[0])
expert_reviews['Reviewed_by'] = 'expert'
expert_reviews.head()
```

```python
## expert and customer reviews have different column names for the same feature
## so for clarity we are going to make them match 

expert_reviews.rename(columns={"model": "Model", "Expert_rating": "Rating"},inplace=True)
expert_reviews.columns
```

```python
## Expert and Customer reviews have different makes in them, if I want to combine them into one sheet 
## it may be best to have the same makes in both  

# Since customer reviews has all of the makes that expert has plus many many more 
# I am going to keep only the makes that expert has in customer reviews


# makes in expert

makes_tokeep = expert_reviews['Company'].unique()
l=sorted(makes_tokeep)
```

```python
# Now we filter out the makes(company) that are not in expert for customer reviews, and compare to check what we have

k = cust_reviews[cust_reviews['Company'].isin(makes_tokeep)]
m = sorted(k['Company'].unique())
compare = pd.DataFrame({'col1':l,'col2':m})
compare
```

```python
# it looks good so we apply the filtered to our main dataframe for customer reviews

cust_reviews = k
cust_reviews['Company'].unique()
```

```python
# So, we continue preparing to  be able to just append customer and expert review together

# expert reviews does not have a 'Date'[date that it was reviewed on] column 
# so we will assume: expert review happened the year of the model of the car 

m_d='01/01/'
new_dates =[]
years = expert_reviews['Year'].astype('str')

for x in years:
    new_dates.append(m_d+x)
new_dates[0:5]    
    
expert_reviews['Date'] = new_dates
expert_reviews.head()
```

```python
# Columns have different position in customer vs expert so we change them to match

expert_reviews = expert_reviews.reindex(columns=['Company', 'Model', 'Year', 'Date', 'Rating', 'Symbol', 'Reviewed_by'])
expert_reviews.head()
```

```python
# customer reviews are in a 5 scale so we can multiply to match

cust_reviews['Rating'] = cust_reviews['Rating']*2
cust_reviews.head()
```

```python
## As you can see above they are looking like they are matching so 
## We can append

review = expert_reviews.append(cust_reviews)
review.head(),review.tail()
```

```python
# There are two issues with Date column to fix:
##  1 the format is not datetime
##  2 there are invalid dates

review['Date'].value_counts(),review.dtypes
```

```python
## make Date column into date format

review["Date"] = pd.to_datetime(review["Date"],errors='coerce')

review["Date"].count(),review.dtypes
```

```python
## how many na's are there? and let's see what they look like

## Because we did the column in the step above, 
## we know that expert reviews has no missing values in date 

review["Date"].isna().sum(),review[review["Date"].isna()].head()
```

```python
# the invalid dates have been converted into NaT, which we can drop (all of them from cust_review)

review.dropna(inplace=True)
review['Date'].sort_values()
```

```python
# Change year to year of ratings for each make(Company)

h = review.groupby(['Symbol',review['Date'].dt.year]).mean()
review_delta_perc = h.groupby('Symbol').pct_change()*100
review_delta_perc.fillna(0, inplace=True)
graph_delta_review = review_delta_perc.reset_index()
graph_delta_review.rename(columns={'Rating':'Change_%'}, inplace=True)
graph_delta_review['Date'] = graph_delta_review['Date'].astype('int64')
graph_delta_review.iloc[17:24]
```

```python
## Change year to year of stock price for each make(Company)

j = stocks.groupby(['Symbol',stocks['Date'].dt.year]).mean()
stocks_delta_perc = j.groupby('Symbol').pct_change()*100
stocks_delta_perc.fillna(0,inplace= True)
graph_delta_stock = stocks_delta_perc.reset_index()
graph_delta_stock['Symbol'].astype('str')
graph_delta_stock.rename(columns={'Open':'Change_%'}, inplace=True)
graph_delta_stock.iloc[37:43]
```
